## 7. Eigenvectors of the Gaussian Covariance Matrix

```python
np.random.seed(42)

rng = np.random.default_rng
sample_size = 100

##############################
### 7.1
X1 = np.random.normal(3, 9**0.5, sample_size)
X2 = 0.5*X1 + np.random.normal(4, 4**0.5, sample_size)
Y = np.vstack((X1, X2)) # 2x100
hat_mu = np.mean(Y, axis=1)
print("7.1\n",f"hat_mu = {hat_mu}\n") 

##############################
### 7.2
centered_Y = Y - hat_mu[:, np.newaxis] # 2x100
hat_cov= (1/sample_size) * centered_Y @ np.transpose(centered_Y)
print("7.2:\n",f"hat_sigma = {hat_cov}\n")

##############################
### 7.3
eigenvalues, eigenvectors = np.linalg.eig(hat_cov)
print("7.3:\n",f"eigenvalues = {eigenvalues}\n")
print(f"eigenvectors = {eigenvectors}\n\n")

##############################
### 7.4

fig6= plt.figure(layout='tight')

##############################
## 7.4(i) Sample Points
plt.scatter(X1, X2, marker=".", color = 'orange', label="Sample Points")

## 7.4(ii) Eigenvectors
plt.quiver(hat_mu[0], hat_mu[1], eigenvalues[0]*eigenvectors[0, 0], eigenvalues[0]*eigenvectors[1, 0], angles='xy', scale_units='xy', scale=1, color='r', label="Eigenvector 1")
plt.quiver(hat_mu[0], hat_mu[1], eigenvalues[1]*eigenvectors[0, 1], eigenvalues[1]*eigenvectors[1, 1], angles='xy', scale_units='xy', scale=1, color='b', label="Eigenvector 2")

plt.title("7.4 Sample Points and Covariance Eigenvectors")
plt.gca().set_aspect('equal', adjustable='box')
plt.xlim(-15, 15)
plt.ylim(-15, 15)

plt.xlabel("$X_1$")
plt.ylabel("$X_2$")
# plt.axvline(x=np.mean(X1), color='b', linestyle='dashed', linewidth=1, label="Mean $X_1$")
# plt.axhline(y=np.mean(X2), color='r', linestyle='dashed', linewidth=1, label="Mean $X_2$")
plt.legend(loc= "upper left")

##############################
### 7.5
fig7= plt.figure(layout='tight')
idx = eigenvalues.argsort()[::-1]   
eigenvalues = eigenvalues[idx]
U = eigenvectors[:,idx]
Y_rotated = U.T @ centered_Y

plt.scatter(Y_rotated[0], Y_rotated[1], marker=".", color = 'orange', label="Sample Points")

plt.title("7.5 Rotated Sample Points in Eigenbasis")
plt.gca().set_aspect('equal')
plt.xlim(-15, 15)
plt.ylim(-15, 15)

plt.xlabel("Eigenvector 1")
plt.ylabel("Eigenvector 2")

plt.show()
```

## 8. Gaussian Classifiers for Digits and Spam

```python
### Partition the data into training and validation sets
mnist = np.load('../data/mnist-data-hw3.npz')
mnist_original_data = mnist['training_data']
mnist_original_labels = mnist['training_labels']
mnist_test_data = mnist['test_data']
print ("mnist_original_data.shape", mnist_original_data.shape)

# Partition the data into training and validation sets
mnist_train_data, mnist_val_data, mnist_train_labels, mnist_val_labels = train_test_split(mnist_original_data, mnist_original_labels, test_size=(1/6), random_state=42)
print ("mnist_train_data.shape", mnist_train_data.shape)
```

### 8.1 Fit a Gaussian distribution to each digit class
### AND
### 8.2 Visualize the covariance matrix for a particular class (digit)
(maximum likelihood estimation w/ pixel values as features)

```python
##############################
### 8.1 
## Contrast-normalize the images before using their pixel values
#  I decided to standardize the normal distribution of the pixel values in the training set rather than divide by the L2 norm of the pixel values since we already assuming our data is distributed Gaussianly.

mnist_train_data = mnist_train_data.astype(np.float32)
mnist_train_data_flattened = mnist_train_data.reshape(mnist_train_data.shape[0], -1)
epsilon = 1e-8

## Standardize pixel values by subtracting the mean and dividing by the standard deviation of all pixels across all training images
# Note: originally tried normalizing by mean/std of each pixel value, but this led to a lot 0 values in the covariance matrix
global_mean = np.mean(mnist_train_data_flattened)
global_std = np.std(mnist_train_data_flattened)
mnist_train_stdized = ((mnist_train_data_flattened - global_mean) / global_std) + epsilon

digits = {}
mu = []
sigma = []

for i in range(10):
    digit_indices= np.where(mnist_train_labels == i)[0]
    digits[i] = mnist_train_stdized[digit_indices]
    # print(f"digits[{i}].shape", digits[i].shape)
    mu.append(np.mean(digits[i], axis=0))
    sigma.append(np.cov(digits[i], rowvar=False))


##############################
### 8.2 
num = 5
plt.figure(figsize=(8, 6))
plt.imshow(sigma[num], cmap='viridis', interpolation='nearest')
plt.title('Covariance Matrix for Digit {}'.format(num))
plt.colorbar()
tick_marks = np.arange(len(sigma[num]))
plt.show()

##############################
```

### 8.3 Classify the digits in the test set on the basis of posterior probabilities 

Code for part a and d as well as b and d are mixed together. Separating "##############################" for clarity.

I used $\tt{scipy.train\_test\_split}$ to partition the data. Because I used the same random seed for each test train split iteration, the same 10,000 randomly chose training points were held out. 


a)  Linear discriminant analysis (LDA): Model P(X|Y=digit)

```python
### 8.3a
np.random.seed(42)
train_nums = [100, 200, 500, 1000, 2000, 5000, 10000, 30000, 50000]
epsilon = 1e-4

overall_errors = []

##############################
### 8.3d  Plot digit-wise error rates
# Dictionary to store digit-wise accuracies for each training size
digit_errors = {i: [] for i in range(10)}
##############################

for train_num in train_nums:
    mnist_train_data, mnist_val_data, mnist_train_labels, mnist_val_labels = train_test_split(mnist_original_data, mnist_original_labels, train_size=train_num, test_size=10000, random_state=42)   

    mnist_train_data_flattened = mnist_train_data.reshape(mnist_train_data.shape[0], -1).astype(np.float32)
    mnist_val_data_flattened = mnist_val_data.reshape(mnist_val_data.shape[0], -1).astype(np.float32)

    ## Standardize pixel values by subtracting the mean and dividing by the standard deviation of all pixels across all training images
    global_mean = np.mean(mnist_train_data_flattened)
    global_std = np.std(mnist_train_data_flattened)
    mnist_train_stdized = ((mnist_train_data_flattened - global_mean) / global_std)
    mnist_val_stdized = ((mnist_val_data_flattened - global_mean) / global_std) 


    digits = {}
    mu = []
    sigma = []
    priors = []

    for i in range(10):
        digit_indices = np.where(mnist_train_labels == i)[0]
        digits[i] = mnist_train_stdized[digit_indices]
        mu.append(np.mean(digits[i], axis=0))
        sigma.append(np.cov(digits[i], rowvar=False))
        priors.append(digits[i].shape[0] / mnist_train_data.shape[0])

    avg_sigma = np.zeros_like(sigma[0]) + epsilon*np.eye(sigma[0].shape[0]) # Add epsilon to avoid singularity (my strategy mentioned in 5.2)
    for i in range(10):
        avg_sigma += priors[i] * sigma[i]

    pdfs = [scipy.stats.multivariate_normal.logpdf(x=mnist_val_stdized, mean=dig_mu, cov=avg_sigma) for dig_mu in mu]

    posteriors = [pdf + np.log(prior) for pdf, prior in zip(pdfs, priors)]
    predictions = np.argmax(posteriors, axis=0)

    for digit in range(10):
        digit_accuracy = np.mean(predictions[mnist_val_labels == digit] == mnist_val_labels[mnist_val_labels == digit])
        digit_errors[digit].append(1 - digit_accuracy)  

    error_rate = 1 - np.mean(predictions == mnist_val_labels) 
    overall_errors.append(error_rate) 

plt.figure(figsize=(6, 4))
plt.plot(train_nums, overall_errors)
plt.xlabel('# of training points')
plt.ylabel('Validation Error rate')
plt.title('LDA classification, overall')
plt.grid(True)
plt.show()

##############################
### 8.3d (cont.) Plot digit-wise error rates
plt.figure(figsize=(6, 4))

for digit in range(10):
    plt.plot(train_nums, digit_errors[digit], label=f'Digit {digit}')

plt.xlabel('# of training points')
plt.ylabel('Validation Error rate')
plt.title('LDA classification, digitwise')
plt.legend()
plt.grid(True)
plt.show()
```

b) Quadratic discriminant analysis (QDA).

```python
### 8.3b
np.random.seed(42)
train_nums = [100, 200, 500, 1000, 2000, 5000, 10000, 30000, 50000]
epsilon = 1e-4 ## any number <=1e-8 is too small (singularity is not avoided)

overall_errors = []

##############################
### 8.3d  Plot digit-wise error rates
# Dictionary to store digit-wise accuracies for each training size
digit_errors = {i: [] for i in range(10)}
##############################

for train_num in train_nums:
    mnist_train_data, mnist_val_data, mnist_train_labels, mnist_val_labels = train_test_split(mnist_original_data, mnist_original_labels, train_size=train_num, test_size=10000, random_state=42)   

    mnist_train_data_flattened = mnist_train_data.reshape(mnist_train_data.shape[0], -1).astype(np.float32)
    mnist_val_data_flattened = mnist_val_data.reshape(mnist_val_data.shape[0], -1).astype(np.float32)

    ## Standardize pixel values by subtracting the mean and dividing by the standard deviation of all pixels across all training images
    global_mean = np.mean(mnist_train_data_flattened)
    global_std = np.std(mnist_train_data_flattened)
    mnist_train_stdized = ((mnist_train_data_flattened - global_mean) / global_std)
    mnist_val_stdized = ((mnist_val_data_flattened - global_mean) / global_std) 

    digits = {}
    mu = []
    sigma = []
    priors = []

    for i in range(10):
        digit_indices = np.where(mnist_train_labels == i)[0]
        digits[i] = mnist_train_stdized[digit_indices]
        mu.append(np.mean(digits[i], axis=0))
        sigma.append(np.cov(digits[i], rowvar=False) + epsilon*np.eye(mu[i].shape[0])) # Add epsilon to avoid singularity (my strategy mentioned in 5.2)
        priors.append(digits[i].shape[0] / mnist_train_data.shape[0])

    pdfs = [scipy.stats.multivariate_normal.logpdf(x=mnist_val_stdized, mean=mu[i], cov=sigma[i], allow_singular=True) for i in range(10)]

    posteriors = [pdf + np.log(prior) for pdf, prior in zip(pdfs, priors)]
    predictions = np.argmax(posteriors, axis=0)

    for digit in range(10):
        digit_accuracy = np.mean(predictions[mnist_val_labels == digit] == mnist_val_labels[mnist_val_labels == digit])
        digit_errors[digit].append(1 - digit_accuracy)  

    error_rate = 1 - np.mean(predictions == mnist_val_labels) 
    overall_errors.append(error_rate) 

plt.figure(figsize=(6, 4))
plt.plot(train_nums, overall_errors)
plt.xlabel('# of training points')
plt.ylabel('Validation Error rate')
plt.title('QDA classification, overall')
plt.grid(True)
plt.show()

##############################
### 8.3d (cont.) Plot digit-wise error rates
plt.figure(figsize=(6, 4))

for digit in range(10):
    plt.plot(train_nums, digit_errors[digit], label=f'Digit {digit}')

plt.xlabel('# of training points')
plt.ylabel('Validation Error rate')
plt.title('QDA classification, digitwise')
plt.legend()
plt.grid(True)
plt.show()
```
